```python
import open3d as o3d

# Создание примерного облака точек
points = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
point_cloud = o3d.geometry.PointCloud()
point_cloud.points = o3d.utility.Vector3dVector(points)

# Создание воксельной сетки с размером вокселя 1.0
voxel_size = 1.0
voxel_grid = o3d.geometry.VoxelGrid.create_from_point_cloud(point_cloud, voxel_size)

# Визуализация воксельной сетки
o3d.visualization.draw_geometries([voxel_grid])

```

```python
import open3d as o3d
import numpy as np
from collections import defaultdict

# Создание облака точек
points = np.random.rand(100, 3) * 10  # 100 случайных точек в диапазоне 0-10
point_cloud = o3d.geometry.PointCloud()
point_cloud.points = o3d.utility.Vector3dVector(points)

# Размер вокселя
voxel_size = 1.0

# Функция для вычисления индексов вокселя
def compute_voxel_index(point, voxel_size):
    voxel_index = tuple(np.floor(point / voxel_size).astype(int))
    return voxel_index

# Подсчет количества точек в каждом вокселе
voxel_count = defaultdict(int)

for point in points:
    voxel_idx = compute_voxel_index(point, voxel_size)
    voxel_count[voxel_idx] += 1

# Печать количества точек в каждом вокселе
for voxel, count in voxel_count.items():
    print(f"Воксель {voxel} содержит {count} точек")

# Создание воксельной сетки из облака точек для визуализации
voxel_grid = o3d.geometry.VoxelGrid.create_from_point_cloud(point_cloud, voxel_size)

# Визуализация воксельной сетки
o3d.visualization.draw_geometries([voxel_grid])
```
